### Number of infections per day analysis

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
corona_dataset=pd.read_csv("covid19_Confirmed_dataset.csv")
```

```python
corona_dataset
```

```python
corona_dataset.shape
```

```python
len(corona_dataset)
```

```python
corona_dataset.head()
```

```python
corona_dataset.drop(["Lat","Long"],axis=1,inplace=True)
```

```python
corona_dataset
```

#### Aggregating the rows by the country

```python
corona_dataset_agg=corona_dataset.groupby("Country/Region").sum()
```

```python
corona_dataset_agg
```

```python
corona_dataset_agg.shape
```

```python
corona_dataset.shape
```

```python
corona_dataset_agg.loc["India"][:9].plot()
```

```python
corona_dataset_agg.loc["India"].plot()
corona_dataset_agg.loc["China"].plot()
corona_dataset_agg.loc["United Kingdom"].plot()
corona_dataset_agg.loc["US"].plot()
plt.legend()
```

```python
corona_dataset_agg.loc["India"].diff().plot()
```

#### Calculating the maximum infection rate in each country

```python
corona_dataset_agg.loc["India"].diff().max()
```

```python
countries=list(corona_dataset_agg.index)
infect_rate_at_max=[]
for each in countries :
    infect_rate_at_max.append(corona_dataset_agg.loc[each].diff().max())
```

```python
countries
```

```python
infect_rate_at_max
```

```python
corona_dataset_agg["infect_rate_at_max"]=infect_rate_at_max
```

```python
corona_dataset_agg.head()
```

```python
final_ansis=pd.DataFrame(corona_dataset_agg["infect_rate_at_max"])
```

```python
final_ansis
```

### Death rate analysis

```python
deaths=pd.read_csv("covid19_deaths_dataset.csv")
```

```python
deaths.head()
```

```python
deaths.shape
```

```python
deaths.drop(["Lat","Long"],axis=1,inplace=True)
```

```python
deaths
```

```python
deaths_data_agg=deaths.groupby("Country/Region").sum()
```

```python
deaths_data_agg
```

```python
deaths_data_agg.shape
```

```python
deaths_data_agg.loc["India"].plot()
deaths_data_agg.loc["China"].plot()
deaths_data_agg.loc["United Kingdom"].plot()
deaths_data_agg.loc["US"].plot()
plt.legend()
```

```python
deaths_data_agg.loc["India"].plot()
```

```python
deaths_data_agg.loc["India"].diff().plot()
```

```python
deaths_data_agg.loc["India"].diff().max()
```

```python
countries_death=list(deaths_data_agg.index)
death_rate_at_max=[]
for eachDeath in countries_death :
    death_rate_at_max.append(deaths_data_agg.loc[eachDeath].diff().max())
```

```python
death_rate_at_max
```

```python
deaths_data_agg["death_rate_at_max"]=death_rate_at_max
```

```python
deaths_data_agg.head()
```

```python
death_ansis=pd.DataFrame(deaths_data_agg["death_rate_at_max"])
```

```python
death_ansis.head()
```

```python
death_join=death_ansis.join(happy_report,how="inner")
```

```python
death_join
```

```python
death_join.shape
```

```python
death_join.corr()
```

#### plotting GDP vs maximum death rate

```python
x=death_join["GDP per capita"]
y=death_join["death_rate_at_max"]
sns.scatterplot(x,np.log(y))
```

```python
sns.regplot(x,y)
```

#### plotting social support vs maximum death rate

```python
x=death_join["Social support"]
y=death_join["death_rate_at_max"]
sns.regplot(x,y)
```

#### plotting healthy life expectancy vs maximum death rate

### plotting freedom to make life choices vs maximum death rate
